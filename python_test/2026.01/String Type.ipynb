```python
# 주민번호 앞 2자리를 생년으로 추출 year
# 주민번호 3~4 자리를 생월로 추출하고 month
# 주민번호 5~6 자리를 생일로 추출 day
# 주민번호 8번째 1자리를 남여 추출 sex

month=inputSsn[2:4]
day=inputSsn[4:6]
sex=inputSsn[7]
print(inputSsn)
print("귀하의 생년은 " + year + "년")
print("귀하의 생월은 " + month + "월")
print("귀하의 생일은 " + day + "일 입니다.")

ssnNumber =inputSsn[7]
#print(ssnNumber)
#type(ssnNumber)
ssnNumber = int(ssnNumber)
#type(ssnNumber)

if ssnNumber %2 == 1:
    print("남성")
else:
    print("여성")
```
