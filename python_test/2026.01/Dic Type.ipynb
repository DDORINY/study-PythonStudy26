```python
# 딕셔너리 타입은 key:value로 이루어진다.
#{ key1:value1, key2:value2,key3:value3....}
# 대부분 웹 프로그램 언어는 json이라는 값으로 처리하는데 유사한 형태이다.

dic ={"name":"kkw", "age": 40}
#        k     v      k     v
type(dic)
print(dic["name"])
#          key
```

```python
a ={1:'hi'}
print(a[1])
#       key
```

```python
b = {"a":[1,2,3]}
#     K     v
print(b["a"])
```

```python
print(b["a"][1])
```

```python
# dic 변수에 email key를 갖는 값을 추가하는 방법
dic['email'] ="kkw@mbc.com"
print(dic)
```

```python
# key 리스트를 생성하려면 a.keys()
dic.keys()
```

```python
# key의 값들을 리스트로 생성하려면
dic.values()
```

```python
# key와 value의 쌍으로 튜플처리하면
dic.items()
```

```python
# get을 이용해 키를 넣으면 값이 나오도록
dic.get("name")
```

```python
# key에 해당하는 값을 가져올 때 없는 키를 넣으면 오류발생?
dic["age"] #없는 키를 가져오려면 오류발생 key Error
# 이것은 프로그램 실행 중 치명적인 오류로 강제 종료된다.
```

```python
dic.get("age","정보없음")
#         k     k가 없을 때 출력문
```
