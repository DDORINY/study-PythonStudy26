```python
# 만약 입력값이 몇개가 될지 모를때는?
def add_many(*args): # *을 붙이면 몇개의 값이 들어오든!
    # *의 뜻은 와일드 문자로 예)김씨로 시작한다. "김*"
    # "*원"이면은 원으로 끝나는 문자라는 뜻이다.
    # "*딸기*" 딸기라는 글자가 포함된이라는 뜻이다.
    # *args ->인수라는 뜻으로 다른 글자를 사용해도 됨/단, 프로그램 관례로 대부분 args로 사용함
    result=""
    for i in args: # args는 튜플로 들어옴
        result = result + i #튜플로 되어 있는 값을 번갈아 더함
    return result
```

```python
# add_many 테스트
result = add_many(1,2,3,6)
print(result)
```

```python
resultText = add_many("안녕하세요","반갑습니다.","파이썬 지옥입니다.")
print(resultText)
```

```python
# 변수선언에는 전역변수가 있고 지역변수가 있다.
# 전역변수는 함수나 if,for, while문 바깥쪽! 맨 뒤에 작성하는 것이 좋다.
# 지역변수는 함수나 if,for, while문에서 1회용으로 사용되는 것으로 안쪽에 선언한다.
```

```python
g1 = 1
def vartest(g1):
    g1 +=1
    return g1 # 변수에 리턴을 넣으면 받을 수 있는 형태

g1 = vartest(g1) # 함수 실행 후 결과를 g1에 다시 넣음
print(g1)
```

```python
a = None # 전역변수 a
def vartest1(a): # 지역변수 a    ===========서로 다른 a변수 이다.
    a = a+1
    # 리턴없음
vartest1(3)
print(a) # a라는 변수는 로컬 변수임으로 def가 종료되면 사라짐
```

```python
a = None
def vartest2(a):
    a = a+1
    return a # 위 코드와 변화된 내용은 return과
a=vartest2(3) # 함수 처리 후 결과를 a에 넣음
print(a)
```
