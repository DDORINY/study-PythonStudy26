```python
# 리스트에  C R U D 기법을 적용해보자!
# Create -> 생성, 추가, 삽입 (.append())
# Read -> 전체리스트 보기/한 개의 자료 보기(.index())
# update -> 자료수정 (="4486")
# Delete -> 자료삭제 (del passwords[3]) /.remove("4486") .pop(3)
```

```python
# 리스트에 자료를 추가해보자!!!
ids = ["kkw","lhj","ksb"] # 회원 id
ids.append("MBC") #회원가입을 id 또는 아이디 등을 생성할때는 변수.append를 사용하여 추가한다.
print(ids)
```

```python
passwords = ["1234","4321","4567"]
passwords.append("8888")
print(passwords)
```

```python
# 리스트에 있는 값 가져오기
ids.index("MBC")
```

```python
# MBC id에 인덱스를 알아와 패스워드의 값을 찾아온다.
passwords[3]
```

```python
# 알아온 패스워드값을 4486으로 수정해본다.
passwords[3] ="4486"
print(passwords)
```

```python
# 3번째 인텍스에 암호를 삭제해본다.
#el passwords[3]
```

```python
passwords.remove("4486")
print(passwords)
```

```python
# pop은 삭제되는 내용을 출력하고 삭제한다.
# passwords.pop() -> 맨 마직막 자료를 삭제
# passwords.pop(2) -> 2 인덱스를 삭제한다.
passwords.pop(2)
```
