```python
# 파일로 더미데이터 및 추가자료를 관리해본다.
# 지금까지는 변수 선언을 할때 초기값을 리스트에 담아 제공을 받았다.
# 그러나 프로그램 종료하면 추가한 값은 사라진다!!!
# 자료를 영구적으로 보관하는 방법은 파일로 저장던지 데이터베이스에 보관

# 서버 : 24시간 365일 동안 멈춤없이 서비스를 제공하는 컴퓨터
# 서비스 : 웹서비스, 게임서비스, 쇼핑서비스, 은행뱅킹서비스 등이 있다.

# 파일저장 : 텍스트파일(메모장), csv(엑셀), json(웹api) 등으로 저장이 가능하다.
```

```python
f = open("새파일.txt","w")
# open은 파일을 열때 사용
# 옵션 : (r) read 읽기전용 | (w) writer 쓰기가능 | (a) append 추가가능( 마지막행에 추가)
# 파일을 열때 경로를 지정하지 않으면 같은 위치에 파일을 찾거나 쓴다.
f.close()
# 파일은 윈도우 정책에서 단일처리를 한다. (열고 작업이 끝나면 닫고 나와야 함)
# 안 닫으면 다른 프로그램에서 열수 없다.
```

```python
f = open("c:/mbc/새파일.txt","w")
f.close()
# 주의사항 경로에 /를 붙이는게 일반적이고 만약 \을 사용하고 싶으면 [c:\\mbc\\새파일.txt ]을 사용해야한다.
# \t는 탭, \n 엔터 \는 이스케이프 문자로 기능이 있는 문자열이기 때문에 2번을 써야 하나로 인식을 한다.
```

```python
f = open("c:/mbc/새파일.txt","w")
for i in range(1,11): #range (1~10까지의 값을 가지고 있다.)
    date = "%d번째 줄입니다.\n" %i
    f.write(date) #.write() 기록!!
    # 파일에 저장
f.close()
```

```python
# 파일에 있는 내용을 가져와 보자!!!
#.readline() 파일에 있는 내용 줄단위로 가져오기
f = open("c:/mbc/새파일.txt","r") # r은 읽기 전용 모드
line = f.readline() # 기본값은 첫줄 가져오기
print(line)
f.close()
```

```python
# 모든 줄을 읽어 화면에 출력하고 싶으면
f = open("c:/mbc/새파일.txt","r")
while True: # false 처리될때까지 반복문
    line = f.readline()
    if not line :break  #라인이 없으면 멈춤
    print(line)
f.close()
# 메모장 내용에 엔터가 적용이 되어 있는 상태라서 출력값에 엔터가 있음 \n\t 라인피드 캐리지리턴
```

```python
# .readlines() 메모장 파일을 리스트로 한번에 불러온다.
# 파일의 모든 줄을 읽어 리스트 형태로 반환
# 반환값은 문자열 리스트 ['1번째줄\n','2번째줄\n','3번째줄\n',....]
# 파일 내용을 모두 메모리에 로드하기 때문에 파일일크기가 클경우 메모리 부족!
# 장점 : 파일의 모든 줄을 한번에 가져와 처리해야 할 때 편리

f = open("c:/mbc/새파일.txt","r")
lines =f.readlines() # 파일에 있는 모든 내용을 리스트처리한다.
#print(lines)
for line in lines:
    print(line)
f.close()
```

```python
# 메모장에 엔터효과인 \n 을 지우고 싶으면 strip
# strip은 문자열내에서 원하는 문자열 또는 공백을 제거함
# lstrip() 원하는 문자열 또는 공백을 왼쪽에서 제거
# rstrip() 원하는 문자열 또는 공백을 오른쪽에서 제거

f = open("c:/mbc/새파일.txt","r")
lines = f.readlines()
for line in lines:
    line = line.strip()
    print(line)
f.close()
```

```python
# read()를 이용하여 파일 문자열 읽기
# 파일의 내용 전체를 문자열로 반환한다.
f =open("c:/mbc/새파일.txt","r")
data =f.read()
print(data) # for문을 사용하지 않고 전체를 읽어 출력!!
f.close()
```

```python
# for를 이용해 줄 단위로 가져오기
f = open("c:/mbc/새파일.txt","r")
# 파일 객체는 기본적으로 for문과 같이 함께 사용하여 줄단위로 읽기도 가능
for line in f:
    print(line)
f.close()
```

```python
# 파일에 새로운 내용을 추가하려면 append -> a
# w로 파일열면 이미 존재하는 파일이면 그 내용이 사라지고 덮어쓰기가 된다.
# a 모드는 원래 있던 값을 유지하며서 새로운 값을 맨 아래에 추가!!
f = open("c:/mbc/새파일.txt","a") # append 모드 (추가모드)
f.write("-------------------------")
for i in range(11,20):
    data = "%d번째 줄입니다. \n" %i
    f.write(data)
f.close()
```
